```python
'''def get_abs_timestamp(ts):
    t = datetime.strptime(ts, "%Y-%m-%d %H:%M:%S.%f")
    t_suff = datetime.strptime(ts.split()[0], "%Y-%m-%d")
    
    return (t - t_suff).total_seconds()*1000'''
    

def get_frame_annot(frame_nb, suffix_path=suffix_path, annot=annot):
    frames_path = os.path.join(suffix_path, "Frames")
    #frame = cv2.imread(os.path.join(frames_path, str(frame_nb)+".png"))
    frame = util.get_frame(cap, frame_nb)
    
    annot_i = pd.concat([pd.Series(list(annot.loc[frame_nb]["bbox"])), pd.Series(annot.loc[frame_nb]["ID"])], axis=1)
    annot_i.columns = ["bbox", "ID"]
    
    # Convert the datetime to daily relative time in miliseconds
    t = datetime.strptime(annot.loc[frame_nb].timestamp, "%Y-%m-%d %H:%M:%S.%f")
    t_suff = datetime.strptime(annot.loc[frame_nb].timestamp.split()[0], "%Y-%m-%d")
    annot_i["AbsTime [s]"] = (t - t_suff).total_seconds()*1000
    
    annot_i.set_index("ID", inplace=True)
    
    return frame, annot_i
```

```python
def crop_img_rect(pts, img):
    # Option 1 : Get the rotated bounding box
    # rect = cv2.minAreaRect(pts) # ((x, y), (w, h), rotation_angle)
    # im_crop, _ = crop_rect(img, rect)
    
    #x,y,w,h = cv2.boundingRect(cnt)
    [x_min, y_min], [x_max, y_max] = np.min(pts, axis=0), np.max(pts, axis=0)
    im_crop = img[y_min:y_max, x_min:x_max]
    

    # Define 8-tuple with x,y coordinates of top-left, bottom-left, bottom-right and top-right corners and apply
    #w, h = x_max-x_min, y_max-y_min
    #transform=pts.flatten()
    #im_crop = (Image.fromarray(img).transform((w,h), ImageTransform.QuadTransform(transform))).toarray
    
    return im_crop


def crop_rect(img, rect):
    # get the parameter of the small rectangle
    center = rect[0]
    size = rect[1]
    angle = rect[2]
    center, size = tuple(map(int, center)), tuple(map(int, size))

    # get row and col num in img
    rows, cols = img.shape[0], img.shape[1]

    M = cv2.getRotationMatrix2D(center, angle, 1)
    img_rot = cv2.warpAffine(img, M, (cols, rows))
    out = cv2.getRectSubPix(img_rot, size, center)

    return out, img_rot
```

```python
def show_image_inline(img, trans_level=1, figsize=None):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    if figsize:
        plt.figure(figsize=figsize)
    plt.imshow(img, alpha=trans_level)
    plt.show(block=False)
        
    return


# BBOX, CIRCLE, LINE, TEXTS
def draw(frame, objects, objects_type="bbox", color=(0, 0, 255)):
    frame = frame.copy()
    if objects_type == "bbox":
        #frame = cv2.drawContours(frame, objects, 0, color, 3)
        for obj in objects:
            frame = cv2.drawContours(frame, np.array([obj]), 0, color, 3)
    elif objects_type == "circle":
        for obj in objects:
            frame = cv2.circle(frame, obj, radius=0, color=color, thickness=20)
    elif objects_type == "line":
        for obj in objects:
            frame = cv2.line(frame, obj[0], obj[1], color, 3)
    elif objects_type == "text":
        bb_objects, text_objects = objects
        for (bb_obj, txt_obj) in zip(bb_objects, text_objects):
            if len(bb_obj) != 2:
                bb_obj = tuple(map(int, np.mean(bb_obj, axis=0)))
            frame = cv2.putText(frame, str(txt_obj), bb_obj, cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 0), 6, cv2.LINE_AA)
    
    return frame
```

```python
# filter_category : [0, 1] to filter car and bus
def read_robert_annot(fp, filter_category=None):
    with open(fp) as f:
        tmp = f.readlines()

    # Each Entry
    # <frame_number> <vehicle_id> <x_left> <y_top> <width> <height> <category> <confidence_score> <category>
    # Where <category> : 0 (Car) / 1 (Bus) / 2 (Truck) / 3 (Motorcycle)
    tmp = [_.split() for _ in tmp]
    tmp = [[eval(x) for x in _] for _ in tmp]
    tmp = [ [x[0]-1, x[1], np.array([[x[2], x[3]], [x[2]+x[4], x[3]], [x[2]+x[4], x[3]+x[5]], [x[2], x[3]+x[5]]]), x[-2], x[-1]]  for x in tmp]
    tmp_df = pd.DataFrame(tmp, columns=["frame", "ID", "bbox", "category", "confidence"])
    tmp_df = tmp_df.sort_values(by=['frame', 'ID'])
    if filter_category is not None:
        tmp_df = tmp_df.loc[tmp_df['category'].isin(filter_category)]
    tmp_df = tmp_df.groupby(["frame"]).agg(list)
    
    return tmp_df

annot_path = suffix_path + ".pkl"

if not os.path.isfile(annot_path):
    annot_tmp = read_robert_annot(suffix_path + ".txt", filter_category=None)
    drone_tmp = pd.read_csv(suffix_path+".CSV")
    drone_tmp = drone_tmp[drone_tmp["frame"] % 4 == 0]
    
    annot = drone_tmp.merge(annot_tmp, on="frame")
    annot.index = annot.frame
    annot.to_pickle(annot_path)
```

### ID Assignment

```python
# output the IDs of the bbox_to
def naive_tracking(bbox_from, bbox_to, ID_from, id_max):
    bbox_from, bbox_to = map(lambda x: [Polygon(bb) for bb in x], [bbox_from, bbox_to])
    
    bbox_from_gpd = gpd.GeoDataFrame({"geometry": bbox_from, "id": ID_from})
    bbox_to_gpd = gpd.GeoDataFrame({"geometry": bbox_to, "id_tmp": range(len(bbox_to))})
    
    overlay_df = gpd.overlay(bbox_from_gpd, bbox_to_gpd, how='intersection')
    overlay_df = process_overlay_df(overlay_df) # keep only the mapping with the bigger overlapping area
    
    from_to_dict = {i_tmp: i for (i_tmp, i) in overlay_df[["id_tmp", "id"]].values}
    
    ID_to = []
    for i in range(len(bbox_to)):
        tmp_id = -1
        if i in from_to_dict.keys():
            tmp_id = from_to_dict[i]
        else:
            id_max += 1
            tmp_id = id_max
        ID_to.append(tmp_id)
    
    return ID_to, id_max


def process_overlay_df(overlay_df):
    overlay_df_grouped = overlay_df.groupby(by=["id"], axis=0).agg(list)
    
    for to_rm_id in overlay_df_grouped[overlay_df_grouped["id_tmp"].map(len) > 1].index:
        processing_row = overlay_df_grouped.loc[to_rm_id]
        max_area_id = processing_row.id_tmp[np.argmax([_.area for _ in processing_row.geometry])]
        id_tmp_ls = processing_row.id_tmp
        for id_tmp_ in id_tmp_ls:
            if id_tmp_ != max_area_id:
                overlay_df.drop(overlay_df[(overlay_df.id==to_rm_id) & (overlay_df.id_tmp==id_tmp_)].index[0], inplace=True)

    overlay_df.reset_index(drop=True, inplace=True)
    
    return overlay_df
```

```python
class VirtualLoopDetector:
    VLD_coords = []
    VLD_lines = []
    
    start_dict, end_dict = {}, {}
    travel_time = {}
    v_patches_pair = {}
    v_additional_patches = {}
    
    train_v_ids, test_v_ids = [], []
    
    def __init__(self, drone_id, annot):
        if drone_id == 10:
            #self.src_coord, self.dst_coord = [(2525, 720), (2528, 1080)], [(1630, 710), (1621, 950)]
            #self.src_coords, self.dst_coord = [(2197, 720), (2200, 1080)], [(500, 680), (491, 940)]
            self.VLD_coords.append([ [(2200, 720), (2200, 1000)], [(500, 680), (491, 940)] ]) # Right -> Left
            self.VLD_coords.append([ [(2200, 1000), (2290, 1120)], [(1850, 1900), (1990, 1900)] ]) # Right -> Bottom
            self.VLD_coords.append([ [(1900, 1130), (1890, 1370)], [(3550, 1160), (3540, 1400)] ]) # Left -> Right
            self.VLD_coords.append([ [(1800, 1050), (1900, 1130)], [(2170, 200), (2380, 200)] ]) # Left -> Top
            self.VLD_coords.append([ [(2150, 1340), (2320, 1340)], [(2170, 190), (2380, 190)] ]) # Bottom -> Top
            self.VLD_coords.append([ [(2050, 1370), (2150, 1340)], [(490, 680), (481, 940)] ]) # Bottom -> Left
            self.VLD_coords.append([ [(1800, 700), (2000, 700)], [(1850, 1910), (1990, 1910)] ]) # Top -> Bottom
        for [src_c, dst_c] in self.VLD_coords:
            self.VLD_lines.append([LineString(src_c), LineString(dst_c)])
            
        self.start_dict, self.end_dict = self.check_overlaps(annot)
        self.travel_time = self.get_travel_time()
        self.train_v_ids, self.test_v_ids = self.data_ls_train_test_split(ratio=0.7)
        self.v_patches_pair = self.get_v_patches_pair()
        self.v_additional_patches = self.get_v_additional_patches(step=32)
        self.create_reid_dataset()
    
    # show_image_inline(LD.visualize_VLD_on_frame(frame), figsize=(10, 20))
    def visualize_VLD_on_frame(self, img, color=[(255, 0, 0), (0, 255, 0), (0, 0, 255), \
                                                 (255, 255, 0), (255, 0, 255), (0, 255, 255), \
                                                 (255, 125, 0), (125, 255, 0), (255, 0, 125), (125, 0, 255)]):
        img = img.copy()
        #img = draw(img, [self.src_coord, self.dst_coord], objects_type="line", color=color)
        for i, [src_c, dst_c] in enumerate(self.VLD_coords):
            img = draw(img, [src_c, dst_c], objects_type="line", color=color[i])
        return img
    
    def check_overlaps(self, annot):
        print("[INFO] Extracting vehicle pair observed at start VLD & end VLD")
        for f_nb in tqdm(annot.index):
            _, annot_i = get_frame_annot(f_nb)
            for v_id in annot_i.index:
                poly = Polygon(annot_i.loc[v_id].bbox)
                abs_t = annot_i.loc[v_id]["AbsTime [s]"]
                for [src_l, dst_l] in self.VLD_lines:
                    if poly.intersects(src_l) and v_id not in self.start_dict.keys():
                        self.start_dict[v_id] = [f_nb, abs_t]
                    if poly.intersects(dst_l) and v_id in self.start_dict.keys() and v_id not in self.end_dict.keys():
                        self.end_dict[v_id] = [f_nb, abs_t]
        
        return self.start_dict, self.end_dict
    
    def get_travel_time(self):
        print("[INFO] Extracting the travel time between vehicle pair observed at start VLD & end VLD")
        for key in sorted(list(set(self.start_dict.keys()).intersection(set(self.end_dict.keys())))):
            delta_time = self.end_dict[key][1] - self.start_dict[key][1]
            if delta_time > HISTORICAL_NB * TEMPORAL_GRANULARITY:
                self.travel_time[key] = delta_time
            else:
                self.start_dict.pop(key)
                self.end_dict.pop(key)
        
        return self.travel_time
    
    
    def data_ls_train_test_split(self, ratio=0.7):
        random.seed(7)
        
        ls = self.travel_time.keys()

        self.train_v_ids = random.sample(ls, int(len(ls)*ratio))
        self.test_v_ids = list(set(ls) - set(self.train_v_ids))
        
        return self.train_v_ids, self.test_v_ids

    
    def get_v_patch(self, v_id, frame_nb):
        tmp_frame, tmp_annot = get_frame_annot(frame_nb)
        tmp_bbox = tmp_annot.loc[v_id]["bbox"]
        tmp_v_patch = crop_img_rect(tmp_bbox, tmp_frame) if len(tmp_bbox)>0 else None

        return tmp_v_patch


    def align_patch(self, patch, way='horizontal'):
        ratio = patch.shape[0] / patch.shape[1]
        if (way == 'horizontal' and ratio < 1) or (way == 'vertical' and ratio > 1):
            patch = cv2.rotate(patch, cv2.ROTATE__90_CLOCKWISE)

        return patch


    def get_v_patches_pair(self, viz=False):
        print("[INFO] Extracting image patches of the vehicle pair")
        for v_id in tqdm(self.travel_time.keys()):
            tmp1 = self.get_v_patch(v_id, self.start_dict[v_id][0])
            tmp2 = self.get_v_patch(v_id, self.end_dict[v_id][0])
            if tmp1 is not None and tmp2 is not None:
                ratio_1 = tmp1.shape[0] / tmp1.shape[1]
                ratio_2 = tmp2.shape[0] / tmp2.shape[1]
                if ratio_1 < 1 and ratio_2 > 1:
                    tmp2 = self.align_patch(tmp2, way='horizontal')
                elif ratio_1 > 1 and ratio_2 < 1:
                    tmp1 = self.align_patch(tmp1, way='horizontal')

                self.v_patches_pair[v_id] = (tmp1, tmp2)

                if viz:
                    print("ID: ", v_id)
                    show_image_inline(tmp1)
                    show_image_inline(tmp2)

        return self.v_patches_pair

    def get_v_additional_patches(self, step=32, viz=False):
        print("[INFO] Extracting additional image patches of the vehicle pair for the training purpose")
        for v_id in self.train_v_ids:
            # Get every ~1 sec additional snapshot of the vehicle to populate the training dataset per vehicle
            self.v_additional_patches[v_id] = [self.get_v_patch(v_id, _) for _ in range(self.start_dict[v_id][0]+step, self.end_dict[v_id][0], step)]
            
            if viz:
                print(v_id)
                for _ in self.v_additional_patches[v_id]:
                    show_image_inline(_)

        return self.v_additional_patches
    
    def create_reid_dataset(self, save_dir="/home/yura/data/KPNEUMA"):
        print("[INFO] Creating ReID Dataset at %s" % save_dir)
        for v_id in tqdm(self.v_patches_pair.keys()):
            tmp1, tmp2 = self.v_patches_pair[v_id]

            save_train_dir = os.path.join(save_dir, "images_train/")
            if not os.path.exists(save_train_dir):
                os.makedirs(save_train_dir)
            save_query_dir = os.path.join(save_dir, "images_query/")
            if not os.path.exists(save_query_dir):
                os.makedirs(save_query_dir)
            save_gallery_dir = os.path.join(save_dir, "images_gallery/")
            if not os.path.exists(save_gallery_dir):
                os.makedirs(save_gallery_dir)

            if v_id in self.train_v_ids:
                cv2.imwrite(os.path.join(save_train_dir, "%.8d_00_00.png" % v_id), tmp1)
                cv2.imwrite(os.path.join(save_train_dir, "%.8d_01_01.png" % v_id), tmp2)
                v_add_patches = self.v_additional_patches[v_id]
                for i in range(2, 2+len(v_add_patches)):
                    cv2.imwrite(os.path.join("/home/yura/data/KPNEUMA/images_train/%.8d_%.2d_%.2d.png" % (v_id, i, i)), v_add_patches[i-2])
            elif v_id in self.test_v_ids:
                cv2.imwrite(os.path.join(save_query_dir, "%.8d_00_00.png" % v_id), tmp1)
                cv2.imwrite(os.path.join(save_gallery_dir, "%.8d_01_01.png" % v_id), tmp2)
                
        return
    
    def equality_check(self, obj):
        if not (self.start_dict == obj.start_dict \
        and self.end_dict == obj.end_dict \
        and self.travel_time == obj.travel_time \
        and self.train_v_ids == obj.train_v_ids \
        and self.test_v_ids == obj.test_v_ids) :
            return False
        
        np.testing.assert_equal(self.v_patches_pair, obj.v_patches_pair)
        np.testing.assert_equal(self.v_additional_patches, obj.v_additional_patches)
        
        return True
```

```python
def get_distmat_temporal(temporal_model_name, use_gt=False):
    def calculate_tanh_dist(ls, model_loss):
        ls = np.absolute(ls)
        ls = np.array([_ if _ < 2*model_loss else model_loss*2 for _ in ls])
        ls /= model_loss
        ls = np.tanh(ls)
        return ls
    
    query_start_time = [LD.start_dict[int(i)][1] for i in viz_reid.val_pid[:viz_reid.num_query]]
    gallery_end_time = [LD.end_dict[int(i)][1] for i in viz_reid.val_pid[viz_reid.num_query:]]

    if use_gt:
        query_pred_time = [LD.travel_time[int(i)]/1000 for i in viz_reid.val_pid[:viz_reid.num_query]]
        model_loss=1

    else:
        if temporal_model_name == "DNN":
            query_pred_time = get_temporal_prediction("DNN", 3500, test_df['Xs'].values)
            model_loss = np.sqrt(223.615)/10
        elif temporal_model_name == "LSTM":
            query_pred_time = get_temporal_prediction("LSTM", 4000, test_df['Xs'].values)
            model_loss = np.sqrt(126.512)/10
        elif temporal_model_name == "GRU":
            query_pred_time = get_temporal_prediction("GRU", 3000, test_df['Xs'].values)
            model_loss = np.sqrt(101.935)/10
        query_pred_time = [_/10 for _ in query_pred_time]
    
    m, n = len(query_start_time), len(gallery_end_time)
    distmat_temporal = np.ones((m, n))
    for q_i in range(m):
        for g_i in range(n):
            distmat_temporal[q_i, g_i] = gallery_end_time[g_i]/1000 - (query_start_time[q_i]/1000 + query_pred_time[q_i]) # pred = 72 = 7.2 sec -> 7200 milisec
        distmat_temporal[q_i, :] = calculate_tanh_dist(distmat_temporal[q_i, :], model_loss)
    
    return distmat_temporal
```

```python
# encoding: utf-8
class visual_reidentifier:
    device = None
    transform = build_transforms(cfg, is_train=False)
    model = None
    feats = None
    
    train_data, val_data, num_query, num_classes = None, None, None, 576
    train_img, train_pid, train_camid, train_fp = None, None, None, None # train_daata
    val_img, val_pid, val_camid, val_fp = None, None, None, None # val_data
    
    
    def __init__(self):
        cfg.merge_from_file('/home/yura/ReID/reid-strong-baseline/configs/softmax_triplet_with_center.yml')
        cfg.merge_from_list(['MODEL.DEVICE_ID', "('2')", \
                            'DATASETS.NAMES', "('kpneuma')", \
                            'DATASETS.ROOT_DIR', "('/home/yura/data')", \
                            'TEST.NECK_FEAT', "('after')", \
                            'TEST.FEAT_NORM', "('yes')", \
                            'MODEL.PRETRAIN_CHOICE', "('self')", \
                            'TEST.WEIGHT', "('../log_SD_VRAI_/resnet50_model_120.pth')"])
        cfg.freeze()

        if cfg.MODEL.DEVICE == "cuda":
            os.environ['CUDA_VISIBLE_DEVICES'] = cfg.MODEL.DEVICE_ID
        cudnn.benchmark = True

        self.device = torch.device('cuda:0')

        model = build_model(cfg, self.num_classes) # VRAI dataset's train num classes = 576
        model.load_param(cfg.TEST.WEIGHT)
        
        self.model = model.to(self.device)
        self.model = self.model.eval()
        
        '''normalize_transform = T.Normalize(mean=cfg.INPUT.PIXEL_MEAN, std=cfg.INPUT.PIXEL_STD)
        transform = T.Compose([
                    T.Resize(cfg.INPUT.SIZE_TEST),
                    T.ToTensor(),
                    normalize_transform
                ])
        self.transform = transform'''

    # dataset : "small", "large"
    # temp_alpha : [0, 1]
    def load_reid_data(self, dataset, distmat_temporal, temp_alpha=0):
        if dataset == "small":
            base_path = os.path.join(cfg.DATASETS.ROOT_DIR, str.upper(cfg.DATASETS.NAMES))
            #feats = pd.read_pickle("feats_KPNEUMA.pkl")
        elif dataset == "large":
            base_path = os.path.join(cfg.DATASETS.ROOT_DIR, "KPNEUMA_orig")
            #feats = pd.read_pickle("feats_KPNEUMA_orig.pkl")
        
        train_dir_path, query_dir_path, gallery_dir_path = [os.path.join(base_path, _) for _ in ["images_train", "images_query", "images_gallery"]]
        train_img_path, query_img_path, gallery_img_path = [os.listdir(_) for _ in [train_dir_path, query_dir_path, gallery_dir_path]]
        
        pattern = re.compile(r'([0-9]+)_([0-9]+)_([0-9]+).png')
        
        train_pid, train_camid = np.split(np.array([list(map(str, pattern.search(_).groups()))[:2] for _ in train_img_path]), 2, axis=1)
        query_pid, query_camid = np.split(np.array([list(map(str, pattern.search(_).groups()))[:2] for _ in query_img_path]), 2, axis=1)
        gallery_pid, gallery_camid = np.split(np.array([list(map(str, pattern.search(_).groups()))[:2] for _ in gallery_img_path]), 2, axis=1)
        
        train_img_path = [os.path.join(train_dir_path, _) for _ in train_img_path]
        query_img_path = [os.path.join(query_dir_path, _) for _ in query_img_path]
        gallery_img_path = [os.path.join(gallery_dir_path, _) for _ in gallery_img_path]
        
        self.num_query = len(query_img_path)
        self.train_fp, self.val_fp = train_img_path, query_img_path+gallery_img_path
        self.train_pid, self.train_camid = train_pid.flatten(), train_camid.flatten()
        self.val_pid, self.val_camid = np.concatenate([query_pid.flatten(), gallery_pid.flatten()]), np.concatenate([query_camid.flatten(),gallery_camid.flatten()])
        
        feats = self.feature_extractor(self.val_fp)
        #feats = pd.read_pickle("feats_KPNEUMA_orig.pkl")
        #display(feats)
        #print(feats)
        #with open("tmp_feats.txt", "w") as f:
        #    for _ in feats.cpu().numpy():
        #        f.write(str(_)+"\n")
        cmc, mAP = self.evaluate(feats, distmat_temporal, temp_alpha)
        
        return cmc, mAP
        
    def read_imgs(self, img_ls):
        if type(img_ls[0]) is str:
            img_ls = [Image.open(_).convert('RGB') for _ in img_ls]
        else:
            img_ls = [Image.fromarray(_) for _ in img_ls]
        img_ls = [self.transform(_) for _ in img_ls]
        img_ls = [_.unsqueeze(0) for _ in img_ls]
        torch_data = torch.cat(img_ls).to(self.device)
        
        return torch_data
    
    def feature_extractor(self, img_ls, batchsize=64):
        feats = []
        for b in torch.utils.data.DataLoader(img_ls, batch_size=batchsize):
            b_imgs = self.read_imgs(b)
            b_feats = self.model(b_imgs)
            feats.append(b_feats)
        
        self.feats = torch.cat(feats)
        print(self.feats.shape)
        
        return self.feats #.cpu().numpy()
    
    
    def cosine_score(self, img_tuple, batchsize=64):
        feat1, feat2 = self.feature_extractor(img_tuple).cpu().numpy()
        cos_dis = distance.cosine(feat1, feat2)

        return cos_dis

    def evaluate(self, feats, distmat_temporal, temp_alpha):
        # Feature extracted for the whole query dataset and the gallery dataset (all the images are treated in the same time)
        #feats = torch.cat(self.feats, dim=0)
        #if self.feat_norm == 'yes':
            #print("The test feature is normalized")
        feats = torch.nn.functional.normalize(feats, dim=1, p=2)
        # query
        qf = feats[:self.num_query] # query features
        q_pids = np.asarray(self.val_pid[:self.num_query]) # pids of same length as qf (pid for each of the image)
        q_camids = np.asarray(self.val_camid[:self.num_query]) # camids of same length as qf (camid for each of the image)
        # gallery
        gf = feats[self.num_query:] # gallery features
        g_pids = np.asarray(self.val_pid[self.num_query:]) # //
        g_camids = np.asarray(self.val_camid[self.num_query:]) # //
        m, n = qf.shape[0], gf.shape[0] # distance matrix size (betweem each query and gallery image, distance calculated)
        distmat = torch.pow(qf, 2).sum(dim=1, keepdim=True).expand(m, n) + \
                  torch.pow(gf, 2).sum(dim=1, keepdim=True).expand(n, m).t() # all 2 matrice of shape m (33 query) x n (724 gallery)
        # COSINE DISTANCE CALCULATION
        distmat.addmm_(1, -2, qf, gf.t()) # Beta(1) * distmat + Alpha(-2) * (qf @ gf.t()) with qf (33 x 2048) and gf.t() (2048 x 724) -> (33 x 724)
        # If qf[i] and gf[j] is the same feature -> L2 norm^2 == 1 -> distmat[i, j] == 0
        # distmat > 0 Why ?
        distmat = distmat.cpu().detach().numpy()
        distmat *= (1-temp_alpha)
        distmat += temp_alpha * (distmat_temporal)
        cmc, mAP = self.eval_func(distmat, q_pids, g_pids, q_camids, g_camids)

        return cmc, mAP
    
    def eval_func(self, distmat, q_pids, g_pids, q_camids, g_camids, max_rank=50):
        """Key: for each query identity, its gallery images from the same camera view are discarded.
            """
        num_q, num_g = distmat.shape
        if num_g < max_rank:
            max_rank = num_g
            print("Note: number of gallery samples is quite small, got {}".format(num_g))
        indices = np.argsort(distmat, axis=1) # (33, 724) Increasing sorting with index
        matches = (g_pids[indices] == q_pids[:, np.newaxis]).astype(np.int32) # Ordering gallery ids based with increasing distmat values (logical : the smaller the distance, more probable match)
        # g_pids : (724, ) -> g_pids[indices] : (33, 724) / q_pids : (33)
        # match each query image with all the gallery image -> matches : ordered list of gallery vehicle image's ids


        # compute cmc curve for each query
        all_cmc = []
        all_AP = []
        num_valid_q = 0.  # number of valid query
        for q_idx in range(num_q):
            # get query pid and camid
            q_pid = q_pids[q_idx] # vehicle id
            q_camid = q_camids[q_idx] # camera id

            # remove gallery samples that have the same pid and camid with query
            order = indices[q_idx]
            remove = (g_pids[order] == q_pid) & (g_camids[order] == q_camid) # 6 removed here for VERI dataset
            keep = np.invert(remove)

            # compute cmc curve
            # binary vector, positions with value 1 are correct matches
            orig_cmc = matches[q_idx][keep] # orig_cmc = gt(k) # len of the vector = gallery size
            if not np.any(orig_cmc):
                # this condition is true when query identity does not appear in gallery
                continue
            # the following two lines are used to get the first index where the first positive appears
            cmc = orig_cmc.cumsum() # cumsum() : [1, 2, 3] -> [1, 3, 6]
            #print(cmc)
            cmc[cmc > 1] = 1 # : [1, 3, 6] -> [1, 1, 1] -> first matching happened at index 0

            all_cmc.append(cmc[:max_rank])
            num_valid_q += 1.

            # compute average precision
            # reference: https://en.wikipedia.org/wiki/Evaluation_measures_(information_retrieval)#Average_precision
            num_rel = orig_cmc.sum() # To calculate the denominator for the precision calculation
            # orig_cmc = [1, 1, 0, 1, 1] : gt(k) : binary vector
            tmp_cmc = orig_cmc.cumsum() # [1, 2, 2, 3, 3] # To calculate the numerator for the precision calculation
            tmp_cmc = [x / (i + 1.) for i, x in enumerate(tmp_cmc)] # Precision calculation [1/1, 2/2, 2/3, 3/4, 3/5]
            tmp_cmc = np.asarray(tmp_cmc) * orig_cmc # tmp_cmc=P(k) (Precision at cut-off k) & orig_cmc = gt(k)
            AP = tmp_cmc.sum() / num_rel # num_rel=N_gt (Nb of ground truths)
            all_AP.append(AP)

        assert num_valid_q > 0, "Error: all query identities do not appear in gallery"

        all_cmc = np.asarray(all_cmc).astype(np.float32) # shape (33 x 50) : (query_img, max_rank)
        all_cmc = all_cmc.sum(0) / num_valid_q # shape (50) : For each ranking % num_valid_q = 33 (nb of valid query images)
        mAP = np.mean(all_AP)

        return all_cmc, mAP
```

```python
def get_label(v_id, travel_time):
    return travel_time[v_id]


# Current set of Features = [Timestamp (hour, min, sec) as total second value + historical pixel position]
def get_v_feature(v_id, start_dict, historical_nb=0):
    def _get_v_position(v_id, annot):
        #feat = annot[annot.ID == v_id][["x_img [px]", "y_img [px]"]].values[0].tolist()
        feat = list(map(int, annot.loc[v_id]["bbox"].mean(axis=0)))
        return feat
    
    def _get_v_timestamp(v_id, annot):
        feat = annot[annot.ID == v_id]["AbsTime [s]"].values[0]
        return feat
    
    # based on y pixel position
    # Output : lane 1, 2, 3, 4 : leftmost lane == 1
    def _get_manual_lane(y_pos):
        if y_pos > 1180:
            return 4
        elif y_pos > 1157:
            return 3
        elif y_pos > 1128:
            return 2
        else:
            return 1
        
    def _get_v_lane(v_id, annot):
        feat = _get_manual_lane(annot[annot.ID == v_id]["y_img [px]"].values[0])
        return feat
    
    frame_nb = start_dict[v_id][0] #get_frame_nb(start_dict[v_id])
    
    #_, annot = get_frame_annot("%05d" % (frame_nb))
    v_feat = [] #[_get_v_timestamp(v_id, annot), _get_v_lane(v_id, annot)]
    
    for i in range(historical_nb):
        if frame_nb-i*TEMPORAL_GRANULARITY >= 0:
            _, annot = get_frame_annot(frame_nb-i*TEMPORAL_GRANULARITY)
            if v_id in annot.index:
                v_pos = _get_v_position(v_id, annot)
        v_feat.append(v_pos)
        
    return np.array(v_feat).reshape((-1, 2))

#feat = get_v_feature(60, start_dict, historical_nb=10)
```

```python
batch_size = 10
nepoch = 10000
device = "cuda" if torch.cuda.is_available() else "cpu"

temporal_model_name = 'LSTM' # 'LSTM' 'GRU' 'DNN'

input_path = suffix_path + "_traj_train.pkl" #'../time-prediction/kpneuma/traj_train.pkl'
test_path = suffix_path + "_traj_test.pkl" #'../time-prediction/kpneuma/traj_test.pkl'

input_dim = 2 

dm = NEW_DataManager(input_path, test_path, batch_size)

checkpoint_dir = "/home/yura/ReID/time-prediction/model_pth/"

# criterion = nn.MSELoss() / nn.L1Loss()
def create_temporal_model(temporal_model_name, input_dim=input_dim):
    def set_model(model_type, input_dim):
        if model_type == 'GRU':
            model = GRU(input_dim, 1, 32, 2)
        if model_type == 'DNN':
            model = DNN(input_dim)
        if model_type == 'LSTM':
            model = LSTM(input_dim, 1, 32, 2)
        
        return model
    
    def weight_init(m):
        if isinstance(m, torch.nn.Linear):
            nn.init.xavier_normal_(m.weight)
            nn.init.constant_(m.bias, 0)
        
        return
    
    temporal_model = set_model(temporal_model_name, input_dim) # select the model and set the input dimension
    temporal_model = temporal_model.to(device)
    temporal_model.apply(weight_init) # apply initial weight : TODO check which initialization
    
    return temporal_model

def train_temporal_model(temporal_model, optimizer, criterion, checkpoint_dir, temporal_model_name):
    log_path = os.path.join(checkpoint_dir, temporal_model_name+'_{:02d}.txt'.format(nepoch))
    with open(log_path, "w") as f:
        f.write("[INFO] Starting the training - %s Model\n" % temporal_model_name)
    start_time = time.time()
    for epoch in range(nepoch):
        total_train_loss = []
        total_test_loss = []
        for input_, label, _ in dm.train_loader:
            label = label.to(device)
            input_ = input_.to(device)
            temporal_model = temporal_model.to(device)
            output = temporal_model(input_)
            output = output.type(torch.FloatTensor).to(device)
            output_1 = output.squeeze().reshape(-1)
            label = label.reshape(-1)
            loss = criterion(output_1, label)
            #loss = Variable(loss, requires_grad = True)

            total_train_loss.append(loss.detach().item())

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        '''if epoch % 1000 == 0:
            end_time = time.time() - start_time
            end_time = str(timedelta(seconds=end_time))[:-7]
            print('Time [%s],Epoch:%f/10000, loss:%.3f' %
                (end_time, epoch, np.mean(total_train_loss)))'''

            # writer.add_scalar('train_loss', np.mean(total_train_loss), epoch)
            
        if epoch % 100 == 0:
            with torch.no_grad():
                for test_input, test_label, _ in dm.test_loader:

                    test_input = test_input.to(device)
                    test_label = test_label.to(device)
                    test_output= temporal_model(test_input)
                    test_output = test_output.type(torch.FloatTensor).to(device)
                    test_output_1 = test_output.squeeze().reshape(-1)
                    test_label = test_label.reshape(-1)
                    test_loss = criterion(test_output_1, test_label)
                    total_test_loss.append(test_loss.detach().item())
            print("==============================================")
            print('Train:Epoch:%d/10000, loss:%.3f' % (epoch, np.mean(total_train_loss)))
            print('Test:Epoch:%d/10000, loss:%.3f,' % (epoch, np.mean(total_test_loss)))
            
            with open(log_path, "a") as f:
                f.write("==============================================\n")
                f.write('Train:Epoch:%d/10000, loss:%.3f\n' % (epoch, np.mean(total_train_loss)))
                f.write('Test:Epoch:%d/10000, loss:%.3f\n' % (epoch, np.mean(total_test_loss)))
                
        if epoch % 500 == 0:
            save_path = os.path.join(checkpoint_dir, temporal_model_name+'_{:02d}.pth'.format(epoch))
            torch.save(temporal_model.cpu().state_dict(), save_path)
            temporal_model.to(device)
    
    return


# rewrite load_checkpoint here...
def load_temporal_model(temporal_model_name, checkpoint_dir, nepoch, device):
    temporal_model = create_temporal_model(temporal_model_name)
    checkpoint_path = os.path.join(checkpoint_dir, temporal_model_name+'_{:02d}.pth'.format(nepoch))
    temporal_model.load_state_dict(torch.load(checkpoint_path))
    temporal_model.to(device)

    return temporal_model

def evaluate_temporal_model():
    return

# data = test_file['Xs'].values
def get_temporal_prediction(temporal_model_name, nb_ep, data, max_value=2500):
    data = torch.FloatTensor(np.concatenate(data).reshape(-1,15,2))/max_value
    data = data.to(device)
    temporal_model = load_temporal_model(temporal_model_name, checkpoint_dir, nb_ep, device)
    
    pred = []
    with torch.no_grad():
        for b in torch.utils.data.DataLoader(data, batch_size=64):
            pred.append(temporal_model(data))
            
    return np.concatenate(torch.cat(pred).cpu().numpy()).flatten()
```
